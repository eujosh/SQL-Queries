**Question 5**

Which day of the week had the highest number of cancelled flights?(Note 1 = cancelled , 0 = not cancelled;)


**Options:**
- Monday(1)
- Wednesday(3) 
- Saturday(6) 
- Thursday (4) 

```python
%%sql 

WITH CTE AS
(SELECT DayOfWeek AS 'Day', COUNT(Cancelled) AS 'No_of_Cancelled_Flights'
FROM flights
WHERE Cancelled = 1  
GROUP BY DayOfWeek)

SELECT Day, MAX(No_of_Cancelled_Flights)
FROM CTE
```

**Question 7**

What is the most produced model for the manufacturer "BOEING"?

**Options:**
- 717-200
- 737-7H4
- 757-222
- 737-3H4

```python
%%sql 

WITH CTE AS
(SELECT model, COUNT(model) AS m
FROM planes 
WHERE manufacturer = 'BOEING'
GROUP BY model)

SELECT model, MAX(m) AS 'Most_Produced_BOEING_Model'
FROM CTE
```

**Question 8**

What manufacturer had the highest average delay time (DepDelay + ArrDelay)? 

**Options:**
- PAIR MIKE E
- AERONCA
- AVIAT AIRCRAFT INC
- BOEING OF CANADA LTD

```python
%%sql

WITH CTE AS (
SELECT p.manufacturer, AVG(f.DepDelay + f.ArrDelay) AS 'Average_Delay_Time'
FROM flights AS f
INNER JOIN planes AS p
ON f.TailNum = p.TailNum
)

SELECT manufacturer, MAX(Average_Delay_Time)
FROM CTE
```

**Question 9**

How many planes landed at Los Angeles International Airport? 

**Options:**

```python
%%sql 

SELECT a.airport, COUNT(f.Dest)
FROM flights f
INNER JOIN airports a
ON f.Dest = a.iata
WHERE airport = "Los Angeles International"
```

**Question 10**

Which domestic carrier had the best on-time performance (OTP), where OTP is defined as the rate of on time flights with a 15min buffer on departure and arrival?

**Options:**
- American Airlines Inc
- Hawaiian Airlines Inc
- Comair Inc
- Aloha Airlines Inc

```python
%%sql

WITH CTE AS (
SELECT c.code, c.description, COUNT(c.code) AS 'No_of_Carrier'
FROM flights AS f
INNER JOIN carriers AS c
ON c.code = f.UniqueCarrier
GROUP BY c.code
)

SELECT description, ((CAST(COUNT(ct.code) AS REAL) / CAST(ct.No_of_Carrier AS REAL))) AS OTP
FROM flights AS f
INNER JOIN CTE AS ct
ON f.UniqueCarrier = ct.code
WHERE ArrDelay < 15 
AND DepDelay < 15
GROUP BY ct.code
ORDER BY OTP DESC
```
